# Feature Extraction 
## ORB ( Oriented FAST and Rotated BRIEF)

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('dataset/imgs/ship.jpg')
```

```python
query_image = cv2.imread('dataset/imgs/ship.jpg')
train_image = cv2.imread('dataset/imgs/ship.jpg')

query_image_bw  = cv2.cvtColor(query_image, cv2.COLOR_BGR2GRAY)
train_image_bw  = cv2.cvtColor(train_image, cv2.COLOR_BGR2GRAY)
```

```python
orb = cv2.ORB_create()

queryKeypoints, queryDescriptors = orb.detectAndCompute(query_image_bw, None)
trainKeypoints, trainDescriptors = orb.detectAndCompute(train_image_bw, None)
```

```python
matcher = cv2.BFMatcher_create()
matches = matcher.match(queryDescriptors, trainDescriptors)
```

```python
final_img = cv2.drawMatches(query_image, queryKeypoints, train_image, trainKeypoints, matches[:20], None)
plt.imshow(final_img)
plt.show()
```
