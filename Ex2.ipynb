# Edge Detection Methods

```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt
```

```python
def show_img( title,_img, **kwargs):
    plt.imshow(_img, **kwargs)
    plt.title(title)
    plt.xticks([])
    plt.yticks([])
```

```python
filter_ = [
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
]
filter_ = np.array(filter_)
img = cv2.imread('dataset/imgs/baboon.jpg')

sobel_verti_img = cv2.filter2D(img, -1, filter_)
sobel_hori_img = cv2.filter2D(img, -1, np.flip(filter_.T, axis=0))

plt.subplot(131)
show_img( 'Original', img)

plt.subplot(132)
show_img('Vertical Edges',sobel_verti_img, cmap='gray')


plt.subplot(133)
show_img('Horizontal Edges',sobel_hori_img, cmap='gray')

plt.show()
```

```python
hori_plus_verti = sobel_verti_img + sobel_hori_img

show_img('Hori + Verti', hori_plus_verti, cmap='gray')
```

```python
edges = cv2.Canny(img, 100, 200)

plt.figure(figsize=(10, 6))
plt.subplot(131)
show_img('Original', img)


plt.subplot(132)
show_img('Canny Edge Detection', edges, cmap='gray')



plt.subplot(133)
show_img('Canny Edge Detection (color)', edges)

plt.show()
```
