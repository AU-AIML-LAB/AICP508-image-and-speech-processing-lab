```python
import numpy as np
from scipy import signal
import librosa
import matplotlib.pyplot as plt
```

```python
d_file = "dataset/auds/set1.wav"
d1_file = "dataset/auds/set2.wav"

d, sr = librosa.load(d_file, sr=8000)
d1, sr = librosa.load(d1_file, sr=8000)
```

```python
def plot_sig(d, i=None):
    i and plt.subplot(i)
    t = np.linspace(0, len(d)/sr, len(d))
    plt.plot(t, d, lw=1)
```

```python
plot_sig(d, 211)
plot_sig(d1, 212)
```

```python
y = signal.convolve(d, d1)
plot_sig(y)
```
