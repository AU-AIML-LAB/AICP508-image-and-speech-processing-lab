```python
import matplotlib.pyplot as plt
import librosa.core as lc

import numpy as np
import librosa.display

path = 'dataset/auds/speech.wav'
y, sr = librosa.load(path, sr=8000)

n_fft = 1024
```

```python
plt.plot(y, label="Original File")
```

```python

def plot_spectrum(y, sr, title, n_fft, hop_length, win_length, window):
    mag = np.abs(lc.stft(y, n_fft=n_fft, hop_length=hop_length, win_length=win_length, window=window))
    D = librosa.amplitude_to_db(mag, ref=np.max)
    img = librosa.display.specshow(D, sr=sr, x_axis='s', y_axis='linear')
    plt.title(title)
    return img
```

```python
plt.figure(figsize=(8, 10))
plt.subplot(2, 1, 1)

img = plot_spectrum(y, sr, "Wideband",  n_fft, 10, 40, "hamming")
plt.colorbar(img, format='%+2.0f dB')

plt.subplot(2, 1, 2)

img1 = plot_spectrum(y, sr, "Narrowband", n_fft, 100, 400, "hamming")

plt.colorbar(img1, format='%+2.0f dB')
plt.show()

```
