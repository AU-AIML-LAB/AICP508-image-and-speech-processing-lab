```python
import matplotlib.pyplot as plt
import librosa.core as lc

import numpy as np
import librosa.display

path = 'dataset/auds/speech.wav'
y, sr = librosa.load(path, sr=8000)

n_fft = 1024
```

```python
plt.plot(y, label="Original File")
```

```python
mag1 = np.abs(lc.stft(y, n_fft=n_fft, hop_length=10, win_length=40, window='hamming'))

D1 = librosa.amplitude_to_db(mag1, ref=np.max)

img1 = librosa.display.specshow(D1, sr=sr, x_axis='s', y_axis='linear')

```

```python
mag1 = np.abs(lc.stft(y, n_fft=n_fft, hop_length=100, win_length=400, window='hamming'))

D1 = librosa.amplitude_to_db(mag1, ref=np.max)

img1 = librosa.display.specshow(D1, sr=sr, x_axis='s', y_axis='linear')
```
