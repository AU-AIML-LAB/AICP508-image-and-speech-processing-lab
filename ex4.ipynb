# Fourier Transform 

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread('dataset/imgs/ship.jpg' ,0)
```

```python
img_1 = np.fft.fft2(img)
img_2 = np.fft.fftshift(img_1)
img_5 = np.fft.ifft2(img_2)
img_3 = np.fft.ifftshift(img_2)
img_4 = np.fft.ifft2(img_3)
```

```python
plt.figure(figsize=(14, 5), constrained_layout=False)

plt.subplot(1, 5, 1)
plt.title("Original")
plt.imshow(img, "gray")

plt.subplot(1, 5, 2)
plt.title("Fourier Transform")
plt.imshow(np.log(1+ np.abs(img_1)), "gray")

plt.subplot(1, 5, 3)
plt.title("Centralized")
plt.imshow(np.log(1+ np.abs(img_2)), "gray")
plt.subplot(1, 5, 4)
plt.title("Decentralized")
plt.imshow(np.log(1+ np.abs(img_3)), "gray")

plt.subplot(1, 5, 5)
plt.title("Inverse Transform")
plt.imshow(np.log(1+ np.abs(img_4)), "gray")


plt.show()

```
