# Morphological Operations

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('dataset/imgs/aa.tiff', 0)
kernel= np.ones((3,3), np.uint8)

plt.imshow(image)
plt.title("Original")
plt.show()
```

```python
erosion = cv2.erode(image, kernel, iterations = 1)
dilation = cv2.dilate(image,kernel, iterations = 1)
```

```python
plt.imshow(erosion)
plt.title("erosion")
plt.show()
```

```python
plt.imshow(dilation)
plt.title("Dilation")
plt.show()
```
